1. Напишите скрипт, который будет показывать найдена ли указанная в
скрипте веб-страница на сервере

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
```

```python
for url in ('https://www.youtube.com/watch?v=VMj-3S1tku0&list=PLAqhIrjkxbuWI23v9cThsA9GvCAUhRvKZ', 
            'https://www.youtube.com/roazanas_abracadabra', 
            'aahhhaha'):
    try:
        response = urlopen(url)
        if response.getcode() == 200:
            print(f'Страница {response.url} найдена на сервере')
            
    except ValueError as err:
        print(f'Некорректный URL ({err})')
        
    except HTTPError as err:
        print(f'Такой страницы на сервере нет ({err})')
        
    except URLError as err:
        print(f'Что-то пошло не так ({err})')
```

2. Проверьте есть ли у сайта SSL сертификат с использованием модуля requests

```python
import requests
```

```python
for url in ('https://www.google.com/',
            'http://92.42.102.38/'):
    try:
        response = urlopen(url)
        if response.getcode() != 200:
            raise(URLError)
        if response.url.startswith('https://'):
            print(f'SSL протокол используется для {url}')
        else:
            print(f'SSL протокол не используется для {url}')
            
    except ValueError as err:
        print(f'Некорректный URL ({err})')
        
    except HTTPError as err:
        print(f'Такой страницы на сервере нет ({err})')
        
    except URLError as err:
        print(f'Что-то пошло не так ({err})')
```

6. Извлеките и выведете на экран все header теги со страницы https://en.wikipedia.org/wiki/Main_Page с помощью urlopen и BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
response = urlopen('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(response, 'lxml')

for tag in soup.find_all(re.compile('h[1-6]')):
    print(f'{tag.name}: {tag.text}')
```
